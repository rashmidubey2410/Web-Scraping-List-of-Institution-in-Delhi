```python
import urllib.request as urllib2
```

```python
wiki = "https://en.wikipedia.org/wiki/List_of_institutions_of_higher_education_in_Delhi"
page = urllib2.urlopen(wiki)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup((page), "lxml")
print (soup.prettify())
```

```python
all_links= soup.find_all("a")
for link in all_links:
    print (link.get("href"))
```

```python
table_head= soup.find_all('table')
for head in table_head:
    print (head.get("th"))
```

```python
table_head= soup.find_all('th')
table_head
```

```python
r_table= soup.find('table')
r_table
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]

for row in r_table.find_all("tr"):
    cells = row.find_all("td")
    univ = row.find_all("th")
    if len(cells)==5: #Only extract table body not heading
        A.append(cells[0].find(text=True))
        B.append(univ[0].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(cells[3].find(text=True))
        F.append(cells[4].find(text=True))
       
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Name_of_University'])
df['Location']=B
df['Type']=C
df['Establishement_year']=D
df['Specialization']=E
df['Sources']=F
df
```
